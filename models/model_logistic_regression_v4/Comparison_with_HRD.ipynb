```python
import pandas as pd
```

```python
model_v4_predicted_samples = pd.read_csv('./../model_logistic_regression_v4/predicted_vs_original_labels_model_v4.csv')
hrd = pd.read_excel('./../../original_input_data/genomic_features/hrd_score_hrdetect_all_repositories.xlsx')

hrd.to_csv('hrd_score.csv')
```

```python
hrd_score = pd.read_csv('hrd_score.csv')
```

```python
model_v4_predicted_samples_HRD = pd.merge(model_v4_predicted_samples, hrd_score, on='SAMPLE_ID', how='inner')
model_v4_predicted_samples_HRD = model_v4_predicted_samples_HRD.drop(columns=['Unnamed: 0_y'], axis='columns')
model_v4_predicted_samples_HRD
```

```python
def hrd_status(x):
    if x['HRDetect_output_probablity'] >= 0.7:
        return 1
    else:
        return 0


model_v4_predicted_samples_HRD['HRD_status'] = model_v4_predicted_samples_HRD.apply(lambda x: hrd_status(x), axis=1)
model_v4_predicted_samples_HRD
```
