```python
import pandas as pd
import numpy as np
```

# Import labels from all repositories

```python
r1_labels = pd.read_csv('./../../labels/basis_for_sample_labeling/Repository1_samples_with_labels.csv')
r2_labels = pd.read_csv('./../../labels/basis_for_sample_labeling/Repository2_samples_with_labels.csv')
r3_labels = pd.read_csv('./../../labels/basis_for_sample_labeling/Repository3_samples_with_labels.csv')
```

## Preparation of labels from repository 1

```python
r1_labels.drop(columns='Unnamed: 0', axis='columns')
```

```python
r1_labels_v1 = r1_labels
```

```python
r1_labels_v1['labels_v1'] = r1_labels_v1['label'].replace({'true_responder': 1,'responder_based_on_PFS': 1, 'responder_based_on_SD': 1, 'true_nonresponder': 0, 'nonresponder_based_on_SD': 0,'nonresponder_based_on_PFS': 0})
pd.set_option('display.max_rows', None)

r1_labels_v1.drop(columns=['Unnamed: 0', 'label'], axis='columns')
```

## Preparation of labels from repository 2

```python
r2_labels
```

```python
r2_labels_v1 = r2_labels
r2_labels_v1['labels_v1'] = r2_labels_v1['label'].replace({'true_responder': 1,'responder_based_on_PFS': 1, 'responder_based_on_SD': 1, 'true_nonresponder': 0, 'nonresponder_based_on_SD': 0,'nonresponder_based_on_PFS': 0})
r2_labels_v1.drop(columns=['Unnamed: 0', 'label'], axis='columns')
```

## Preparation of labels from repository 3

```python
r3_labels_v1 = r3_labels
r3_labels_v1['labels_v1'] = r3_labels_v1['label'].replace({'true_responder': 1,'responder_based_on_PFS': 1, 'true_nonresponder': 0})
r3_labels_v1.drop(columns=['Unnamed: 0', 'label'], axis='columns')
```

## Merge all labels v1

```python
labels_v1 = pd.concat([r1_labels_v1, r2_labels_v1, r3_labels_v1], ignore_index=True)
labels_v1 = labels_v1.drop(columns=['Unnamed: 0', 'label'], axis='columns')
labels_v1.to_csv('labels_v1.csv')
```
