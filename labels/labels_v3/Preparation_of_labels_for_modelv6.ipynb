```python
import pandas as pd
import numpy as np
```

# Import labels from all repositories

```python
r1_labels = pd.read_csv('./../../labels/basis_for_sample_labeling/Repository1_samples_with_labels.csv')
r2_labels = pd.read_csv('./../../labels/basis_for_sample_labeling/Repository2_samples_with_labels.csv')
r3_labels = pd.read_csv('./../../labels/basis_for_sample_labeling/Repository3_samples_with_labels.csv')
```

## Preparation of labels from repository 1 - selection of samples with uncertain labels (training) and certain labels (test)

```python
r1_labels.drop(columns='Unnamed: 0', axis='columns')
```

```python
r1_labels_v3 = r1_labels
```

```python
# Training set selection

r1_labels_v3['labels_v3'] = r1_labels_v3['label'].replace({'responder_based_on_PFS': 1, 'responder_based_on_SD': 1, 'nonresponder_based_on_SD': 0,'nonresponder_based_on_PFS': 0})
pd.set_option('display.max_rows', None)
r1_labels_v3.drop(columns=['Unnamed: 0', 'label'], axis='columns')
r1_labels_v3_training = r1_labels_v3[r1_labels_v3['labels_v3'].isin([1, 0])]
r1_labels_v3_training_final = r1_labels_v3_training.drop(columns=['Unnamed: 0', 'label'], axis='columns')

# Testing set selection

r1_labels_v3_test = r1_labels_v3[r1_labels_v3['label'].isin(['true_responder', 'true_nonresponder'])]
r1_labels_v3_test['labels_v3'] = r1_labels_v3['label'].replace({'true_responder': 1, 'true_nonresponder': 0})
r1_labels_v3_test_final = r1_labels_v3_test.drop(columns=['Unnamed: 0', 'label'], axis='columns')
r1_labels_v3_test_final
```

## Preparation of labels from repository 2

```python
r2_labels_v3 = r2_labels
r2_labels_v3['labels_v3'] = r2_labels_v3['label'].replace({'possible_responder': 1, 'possible_nonresponder': 0})
pd.set_option('display.max_rows', None)

r2_labels_v3_training_final = r2_labels_v3.drop(columns=['Unnamed: 0', 'label'], axis='columns')
r2_labels_v3_training_final
```

## Preparation of labels from repository 3

```python
# Training set selection

r3_labels_v3 = r3_labels
r3_labels_v3
r3_labels_v3['labels_v3'] = r3_labels_v3['label'].replace({'responder_based_on_PFS': 1})
r3_labels_v3_training = r3_labels_v3[r3_labels_v3['labels_v3'].isin([1])]
r3_labels_v3_training_final = r3_labels_v3_training.drop(columns=['Unnamed: 0', 'label'], axis='columns')
r3_labels_v3_training_final


# Testing set selection

r3_labels_v3['labels_v3'] = r3_labels_v3['label'].replace({'true_responder': 1, 'true_nonresponder': 0})
r3_labels_v3_test = r3_labels_v3[r3_labels_v3['labels_v3'].isin([1, 0])]
r3_labels_v3_test_final = r3_labels_v3_test.drop(columns=['Unnamed: 0', 'label'], axis='columns')
r3_labels_v3_test_final
```

## Merge all labels - training and testing v3

```python
labels_v3_training = pd.concat([r1_labels_v3_training_final, r2_labels_v3_training_final, r3_labels_v3_training_final], ignore_index=True)
labels_v3_training.to_csv('labels_v3_training_set.csv')
```

```python
labels_v3_test = pd.concat([r1_labels_v3_test_final, r3_labels_v3_test_final], ignore_index=True)
labels_v3_test
labels_v3_test.to_csv('labels_v3_test_set.csv')
```
